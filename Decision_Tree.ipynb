```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/nunnarilabs/ml/master/car_data/car.data',names=['buying','maint','doors','persons','lug_boot','safety','class'])
data
```

```python
data['class'],class_names=pd.factorize(data['class']) #label encoding
print(class_names)
```

```python
data_backup=data.copy()
data_backup.head()
```

```python
data['buying'],_=pd.factorize(data['buying'])
data['maint'],_=pd.factorize(data['maint'])
data['doors'],_=pd.factorize(data['doors'])
data['persons'],_=pd.factorize(data['persons'])
data['lug_boot'],_=pd.factorize(data['lug_boot'])
data['safety'],_=pd.factorize(data['safety'])
data.head()
```

```python
x=data.iloc[:,0:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)#will take continuous values
```

```python
x.columns
```

```python
from sklearn import tree,metrics
dtree=tree.DecisionTreeClassifier(criterion='gini') #we can also use entropy
dtree.fit(x_train,y_train)
```

```python
y_test_pred=dtree.predict(x_test)
```

```python
accuracy=metrics.accuracy_score(y_test,y_test_pred)
print('Train accuracy: {:.2f}'.format(accuracy))
```
